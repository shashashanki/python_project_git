```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
nodes = ['alcohol', 'malic_acid', 'ash', 'alcalinity_of_ash', 'magnesium',
    'total_phenols', 'flavanoids', 'nonflavanoid_phenols',
    'proanthocyanins', 'color_intensity', 'hue',
    'od280/od315_of_diluted_wines', 'proline', 'target'
    ]

edges = [('malic_acid', 'hue'),
    ('ash', 'alcalinity_of_ash'),
    ('magnesium', 'proline'),
    ('total_phenols', 'flavanoids'),
    ('flavanoids', 'od280/od315_of_diluted_wines'),
    ('hue', 'target'),
    ('proline', 'target')
    ]
```

```python
# 
G = nx.DiGraph()  #  (Directed Graph)

# 
G.add_nodes_from(nodes)
#  ()      
G.add_edges_from(
                edges
                )

plt.figure(figsize=(15,10))
pos = nx.circular_layout(G, scale=1, center=None, dim=2)  # plot

# https://networkx.org/documentation/stable/reference/generated/networkx.drawing.nx_pylab.draw_networkx.html#draw-networkx
nx.draw_networkx(G, pos,
                node_size=2000,
                node_color='#85B5FF',
                width=2,
                arrowsize=20,
                )
plt.grid(False)
plt.show()
```

```python
nodes = ['T0',
        'L1',
        '2',
        '3',
        '4',
        '5',
        'L6',
        '7',
        '8',
        '9',
        '10',
        'L11',
        '12',
        '13',
        '14',
        '15',
        'L16',
        '17',
        '18',
        '19',
        '20',
    ]

edges = [('L1','T0'),('L6','T0'),('L11','T0'),('L16','T0')]
for i in nodes:
    if i in ['2','3','4','5']:
        edges.append((i,'L1'))
    if i in ['7','8','9','10']:
        edges.append((i,'L6'))
    if i in ['12','13','14','15']:
        edges.append((i,'L11'))
    if i in ['17','18','19','20']:
        edges.append((i,'L16'))


print(edges)
```

```python
# 
G = nx.DiGraph()  #  (Directed Graph)

# 
G.add_nodes_from(nodes)
#  ()      
G.add_edges_from(
                edges
                )

plt.figure(figsize=(15,10))
pos = nx.kamada_kawai_layout(G, scale=1)  # plot
# pos['T0']=np.array([0,0])

# https://networkx.org/documentation/stable/reference/generated/networkx.drawing.nx_pylab.draw_networkx.html#draw-networkx
nx.draw_networkx(G, pos,
                node_size=2000,
                node_color='#85B5FF',
                width=2,
                arrowsize=20,
                )
plt.grid(False)
plt.show()
```

```python
# 
G = nx.DiGraph()  #  (Directed Graph)

# 
G.add_nodes_from(nodes)
#  ()      
G.add_edges_from(
                edges
                )

# 
for i in nodes:
    if i in ['T0']:
        G.nodes[i]['layer']=1
    elif i in ['L1','L6','L11','L16']:
        G.nodes[i]['layer']=2
    else:
        G.nodes[i]['layer']=3

plt.figure(figsize=(15,10))
pos = nx.multipartite_layout(G, subset_key="layer", align="horizontal")
# pos['T0']=np.array([0,0])

# https://networkx.org/documentation/stable/reference/generated/networkx.drawing.nx_pylab.draw_networkx.html#draw-networkx
nx.draw_networkx(G, pos,
                node_size=2000,
                node_color='#85B5FF',
                width=2,
                arrowsize=20,
                )
plt.grid(False)
plt.show()
```
