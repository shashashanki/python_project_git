```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 乱数シードを設定して再現性を確保します
np.random.seed(0)
data_num = 10

# plt.figure(figsize=(8, 4))  # プロットのサイズを設定します
fig, ax = plt.subplots(1,1, figsize=(8,4))

for i in range(data_num):
    # ランダムウォーク
    data = [0]
    for j in range(100):
        data.append(data[j]+np.random.normal(loc=0.0, scale=1.0))

    # プロットを作成します
    ax.plot(data, color='blue', alpha=0.5, linewidth=2, label=f'sample{i}')  # データをプロットします
    
ax.set_xlabel('time')  # x軸のラベルを設定します
ax.set_ylabel('value')  # y軸のラベルを設定します
ax.set_title('title')  # グラフのタイトルを設定します
# ax.grid(True)  # グリッドを表示します
ax.legend()  # 凡例を表示します

# プロットを表示します
plt.show()
```

```python
# 乱数シードを設定して再現性を確保します
np.random.seed(0)
data_num = 10

# plt.figure(figsize=(8, 4))  # プロットのサイズを設定します
fig, ax = plt.subplots(1,1, figsize=(8,4))

# カラーマップを設定します（例: 'viridis' カラーマップを使用）
cmap = plt.get_cmap('viridis')

for i in range(data_num):
    # ランダムウォーク
    data = [0]
    for j in range(100):
        data.append(data[j]+np.random.normal(loc=0.0, scale=1.0))

    color = cmap(i / (data_num - 1))  # カラーマップから色を取得します
    # プロットを作成します
    ax.plot(data, color=color, alpha=0.7, linewidth=2, label=f'sample{i}')  # データをプロットします
    
ax.set_xlabel('time')  # x軸のラベルを設定します
ax.set_ylabel('value')  # y軸のラベルを設定します
ax.set_title('title')  # グラフのタイトルを設定します
# ax.grid(True)  # グリッドを表示します
ax.legend()  # 凡例を表示します

# プロットを表示します
plt.show()
```
