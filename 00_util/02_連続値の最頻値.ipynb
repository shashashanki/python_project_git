```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde
```

```python
np.random.seed(1)

# ランダムウォーク
data = [0]
for j in range(1000):
    data.append(data[j]+np.random.normal(loc=0.0, scale=1.0))
```

```python
# plt.figure(figsize=(8, 4))  # プロットのサイズを設定します
fig, ax = plt.subplots(1,1, figsize=(8,4))

# プロットを作成します
ax.plot(data, color='blue', alpha=0.5, linewidth=2, label=f'sample')  # データをプロットします
    
ax.set_xlabel('time')  # x軸のラベルを設定します
ax.set_ylabel('value')  # y軸のラベルを設定します
ax.set_title('title')  # グラフのタイトルを設定します
# ax.grid(True)  # グリッドを表示します
ax.legend()  # 凡例を表示します

# プロットを表示します
plt.show()
```

```python
# plt.figure(figsize=(8, 4))  # プロットのサイズを設定します
fig, ax = plt.subplots(1,1, figsize=(8,4))

# プロットを作成します
ax.hist(data, color='blue', alpha=0.5, label=f'sample', bins=20)  # データをプロットします
    
ax.set_xlabel('time')  # x軸のラベルを設定します
ax.set_ylabel('value')  # y軸のラベルを設定します
ax.set_title('title')  # グラフのタイトルを設定します
# ax.grid(True)  # グリッドを表示します
ax.legend()  # 凡例を表示します

# プロットを表示します
plt.show()
```

```python
limmin = min(np.percentile(x, 0.1) for x in data)
limmax = max(np.percentile(x, 99.9) for x in data)
ls = np.linspace(limmin, limmax, 100)

x = data
x = np.array(x)[(x > limmin)&(x < limmax)]
kde = gaussian_kde(x)

fig, ax1 = plt.subplots(1,1, figsize=(8,4))
ax1.plot(ls, kde(ls), label='data')

ax2 = ax1.twinx()
ax2.hist(data, color='blue', alpha=0.5, label=f'sample', bins=20)  # データをプロットします

# 密度が最大の時点を取得
time_max = ls[kde(ls).argmax()]

# 赤い縦線を特定の横軸の値に挿入します
plt.axvline(x=time_max, color='red', linestyle='--', label='max', linewidth=2)

ax1.set_xlim([limmin, limmax])
ax1.legend()
ax1.set_title('data distributions')
plt.show()
```
