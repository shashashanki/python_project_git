```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 
np.random.seed(0)
data_num = 10

# plt.figure(figsize=(8, 4))  # 
fig, ax = plt.subplots(1,1, figsize=(8,4))

for i in range(data_num):
    # 
    data = [0]
    for j in range(100):
        data.append(data[j]+np.random.normal(loc=0.0, scale=1.0))

    # 
    ax.plot(data, color='blue', alpha=0.5, linewidth=2, label=f'sample{i}')  # 
    
ax.set_xlabel('time')  # x
ax.set_ylabel('value')  # y
ax.set_title('title')  # 
# ax.grid(True)  # 
ax.legend()  # 

# 
plt.show()
```

```python
# 
np.random.seed(0)
data_num = 10

# plt.figure(figsize=(8, 4))  # 
fig, ax = plt.subplots(1,1, figsize=(8,4))

# : 'viridis' 
cmap = plt.get_cmap('viridis')

for i in range(data_num):
    # 
    data = [0]
    for j in range(100):
        data.append(data[j]+np.random.normal(loc=0.0, scale=1.0))

    color = cmap(i / (data_num - 1))  # 
    # 
    ax.plot(data, color=color, alpha=0.7, linewidth=2, label=f'sample{i}')  # 
    
ax.set_xlabel('time')  # x
ax.set_ylabel('value')  # y
ax.set_title('title')  # 
# ax.grid(True)  # 
ax.legend()  # 

# 
plt.show()
```
